```python
from tdc.single_pred import Tox
import deepchem
data = Tox(name="LD50_Zhu")
df = data.get_data()
splits = data.get_split()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import autokeras as ak
```

```python
#smiles = ['CC(=O)OC1=CC=CC=C1C(=O)O']

featurizer = deepchem.deepchem.feat.RDKitDescriptors()
f = featurizer.featurize(df.iloc[:,1].to_list())
#print(f[0])

total_size = int(f.shape[0])
train_size = int(total_size * 0.9)

print(train_size)
```

```python
newData = pd.DataFrame(f)
#print(newData)
train_data = newData[:train_size]
test_data = newData[train_size:]

train_data_y = df.iloc[:train_size,2]
test_data_y = df.iloc[train_size:total_size,2]

reg = ak.StructuredDataRegressor(overwrite=True, max_trials=50, metrics=['mae'])
reg.fit(x=train_data, y=train_data_y)
```

```python
predicted_y = reg.predict(test_data)
print(predicted_y)
print(reg.evaluate(x=test_data, y=test_data_y))
```
