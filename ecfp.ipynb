```python
from tdc.single_pred import Tox
import deepchem
data = Tox(name="LD50_Zhu")
df = data.get_data()
splits = data.get_split()

import numpy as np
import pandas as pd
import tensorflow as tf
import autokeras as ak
import padelpy
```

```python

# Example 1: (size = 2048, radius = 4)

featurizer = deepchem.feat.CircularFingerprint(size=2048, radius=4)

features = featurizer.featurize(df.iloc[:,1].to_list())
print(features)
```

```python
#print(features.shape)
total_size = int(features.shape[0])
train_size = int(total_size * 0.9)

newData = pd.DataFrame(features)
#print(newData)
train_data = newData[:train_size]
test_data = newData[train_size:]

train_data_y = df.iloc[:train_size,2]
test_data_y = df.iloc[train_size:total_size,2]
```

```python
reg = ak.StructuredDataRegressor(overwrite=True, max_trials=3, metrics=['mae'])
reg.fit(x=train_data, y=train_data_y)
```

```python
predicted_y = reg.predict(test_data)
print(predicted_y)
print(reg.evaluate(x=test_data, y=test_data_y))
```
